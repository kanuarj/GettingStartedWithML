Gaussian Normal Distribution

So instead of using the Gaussian Scale we can just Standardize the Data with fixating the Mean and Standard Deviation 

This is done using <i>Standard Normal Distrubition</i>.

Using <i>Z-Statistic</i> we can do it.

The formula for Z-Statistic is given by $$ Z = \frac {x - \mu}{\sigma} $$

This converts mean = 0 and standard deviation = 1

Find the proportion of the people whose weight is less 60

So the $$ \mu=70, \sigma=10 $$

So calculating Z-Score ~ $$ Z = \frac {60-70}{10} = -1 $$

This value is then verified with Standard Normal Distribution Table which gives us Z value of -1 as .15866 Proportion people less than 60

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
```

```python
import pandas as pd
```

```python
X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target, name="Target")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state = 100)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_train, y_train)
```

```python
scaled_y_pred = knn.predict(X_test)
```

```python
accuracy_score(y_test, scaled_y_pred)
```
