Frequent Count ~ It is the most occurred value in the overall column.<br>
Frequent Count Imputation ~ Replacement of the frequent count value with the NaN values.

```python
import pandas as pd
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.isna().sum()
```

```python
most_occurred = df.Age.value_counts().index[0]
```

```python
import numpy as np
def most_frequency_value(dataset, column_name, occurred_variable):
    dataset[column_name+'imputed'] = np.where(dataset[column_name].isna(), occurred_variable, dataset[column_name])
```

```python
most_frequency_value(df, 'Age', most_occurred)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
simple_imputer = SimpleImputer(strategy = 'most_frequent')
age_imputed_si = simple_imputer.fit_transform(df[['Age']])
```

```python
df['age_imputed_si'] = age_imputed_si
```
